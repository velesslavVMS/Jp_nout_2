1. Выберите дом клиенту. Клиент хочет дом с видом на набережную(waterfront), как минимум с тремя ванными и с подвалом.
Сколько вариантов есть у клиента?  
У клиента 41 вариант.
- Выведите все варианты таблицей

```python
data[(data['waterfront'] == 1) & (data['bathrooms'] >= 3) & ~(data['sqft_basement'] == 0)]
```

2. Выберите дом клиенту. 
Клиент хочет дом
- либо с очень красивым видом(параметр 'view' - больше) из окна, либо с видом на набережную(waterfront)
- и в очень хорошем состоянии(condition) и год постройки не меньше 1980 года.
При условии, что 'view' попадает в верхний перцентиль 75%, т.е 'view'>=4,  у клиента 21 вариант.
- В какой ценовом диапазоне будут дома?
Дома в ценовом диапазоне от 705 000 до 5300 000.

```python
data['condition'].describe()
```

```python
request = data[((data['view'] == 4) | (data['waterfront'] == 1)) & (data['condition'] == 4) & (data['yr_built'] >= 1980)][['view','waterfront','condition','yr_built', 'price']]
request
```

```python
request.shape[0]
```

```python
request['price'].describe()/1000
```

3. Выберите дом клиенту. Клиент хочет дом
- без подвала
- с двумя этажами
- стоимостью до 150000.
- Какая оценка по состоянию у таких домов в среднем?
Средняя оценка состояния таких домов  = 3.

```python
inqw = data[(data['sqft_basement'] == 0) & (data['floors'] == 2) & (data['price'] <= 150000)][['sqft_basement','floors','price','condition']]
inqw
```

```python
inqw['condition'].mean()
```
